```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
x = np.linspace(0,20,100)
y = np.sin(x) + np.random.rand(100) + np.random.randn(100)
error = np.random.rand(100)
```

```python
plt.errorbar(x,y,yerr=error, fmt='.k')
```

```python
plt.errorbar(x,y, yerr= error, fmt ='o', color ='black', ecolor='grey',elinewidth=1)
```
