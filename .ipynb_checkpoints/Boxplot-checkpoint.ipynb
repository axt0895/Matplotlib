```python
#data to plot in boxplot
Apple =[62,77,44,41,77,69,79,79,75,35,48,81,75,64,72,88,81,56,54,59]
Banana =[64,52,81,45,35,31,91,98,19,91,90,52,29,53,103,16,75,13,33,23]
Peach =[25,29,19,98,52,81,21,61,65,85,12,54,12,56,36,55,35,32,22,82]
Watermelon =[89,75,71,19,88,66,89,99,70,80,89,78,14,29,75,86,79,91,73,90]
Grape =[90,57,76,40,18,88,65,81,58,19,47,89,32,36,43,52,18,58,19,95]

data =[Apple, Banana, Peach, Watermelon, Grape]
```

```python
plt.boxplot(data, vert = False, patch_artist=True, 
            boxprops= dict(facecolor='#A1BFD9', color ='#2077B4'),
            labels=['Apple', 'Banana','Peach','Watermelon','Grape'])
plt.title('Box plot')

plt.show()
```
