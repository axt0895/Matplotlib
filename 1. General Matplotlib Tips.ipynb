```python
%matplotlib inline
import numpy as np
import numexpr
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

from pandas import Series, DataFrame
from IPython.display import Image

np.random.RandomState(42)
sns.set()
```

```python
x = np.linspace(0,10,1000)
```

```python
fig = plt.figure(figsize=(4,3))
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

```python
fig.savefig('GeneralMatplotlib.png')
```

```python
plt.figure(figsize=(4,3))

#Create the 1st panel
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))

#Create the 2nd panel
plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
```
