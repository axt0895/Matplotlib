```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
x = np.linspace(0,10,100)
```

```python
plt.figure(figsize = (6, 4))

plt.show()
```

```python
plt.figure(figsize = (6,4))

plt.subplot(3, 1, 1)
plt.plot(x, np.sin(x))

plt.subplot(3, 1, 2)
plt.plot(x, np.cos(x))

plt.subplot(3, 1, 3)
plt.plot(x, np.tan(x))

plt.gca()
```

```python
figure, axes = plt.subplots(2)

axes[0].plot(x, np.sin(x))
axes[1].plot(x, np.cos(x))
```
