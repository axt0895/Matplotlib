```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,10,100)
y = X ** 2 + X +3
```

```python
plt.title('Quadratic Function')
plt.xlabel('Value of x')
plt.ylabel('Value of y')
plt.xlim(0,10)
plt.ylim(0,100)


plt.plot(X, y, label = 'function')
plt.legend()

plt.show()
```

### prints multiple plots

```python
plt.subplot(1, 3, 1)
plt.plot(X, y, color ='red', linestyle ='dotted')

plt.subplot(1, 3, 2)
plt.plot(X, y, color ='blue', linestyle ='dashed')

plt.subplot(1, 3, 3)
plt.plot(X, y, color = 'green', linestyle ='dashdot')
```

```python
plt.figure(figsize = (6, 8))

plt.subplot(3, 1, 1)
plt.plot(X, y, 'r-')

plt.subplot(3, 1, 2)
plt.plot(X, y, 'b:')

plt.subplot(3, 1, 3)
plt.plot(X, y, 'g--')
```

```python
X = np.linspace(0,5, 50)
y = X ** 3 + X ** 2 + X + 1
```

```python
plt.figure(figsize = (6,6))

plt.subplot(2, 2, 1)
plt.plot(X, y, 'r-')

plt.subplot(2, 2, 2)
plt.plot(X, y, 'b:')

plt.subplot(2, 2, 3)
plt.plot(X, y, 'g--')

plt.subplot(2, 2, 4)
plt.plot(X, y, 'y-.')

plt.show()
```

## Using Figure and Axes Objects

```python
fig, ax = plt.subplots(2,2, figsize = (8,8))
ax[0,0].plot(X, y, 'r--', label ='Cubic Function')
ax[0,0].set_xlabel('Values of X')
ax[0,0].set_ylabel('Values of y')
ax[0,0].set_title('Graph of cubic functions')
ax[0,0].set_xlim(0,5)
ax[0,0].set_ylim(0,160)
ax[0,0].legend()

ax[0,1].plot(X, y, 'b:')

ax[1,0].plot(X, y, 'g-')

ax[1,1].plot(X, y, 'y-.')


plt.show()
```
