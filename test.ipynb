```python
%matplotlib inline

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
X = np.array([1,3,5,7,9,11,13,15])
y = np.array([2,6,7,9,13,14.5,16,21])
```

```python
plt.figure(figsize = (6, 4))

plt.plot(X, y, color = 'red', linestyle = 'solid', label ='Car price')
plt.xlabel('Price')
plt.ylabel('Mileage')
plt.title('Car Price vs Mileage')
plt.xlim(0,15)
plt.ylim(0, 22)

plt.legend()
plt.show()
```

```python
X = np.random.randint(0,100, 50)
y = np.random.randint(0,250, 50)
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
plt.scatter(X, y, label ='Scatter Plot')
plt.xlabel('Age')
plt.ylabel('History')
plt.title('Morse Codes')
plt.xlim(0, 100)
plt.ylim(0,250)

plt.legend(loc ='upper left', fancybox = True)
plt.show()
```
