```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

np.random.RandomState(42)
sns.set()
```

```python
data = np.random.randn(1000)
```

```python
plt.hist(data)
```

```python
plt.hist(data, alpha = 0.8)
plt.show()
```

```python
x = np.random.normal(-2, 1, 1000)
y = np.random.normal(0, 1, 1000)
z = np.random.normal(2, 0.5, 1000)
a = np.random.normal(-4,1, 1000)
kwargs = dict(histtype ='stepfilled', alpha = 0.45, bins = 35)
```

```python
plt.hist(x, **kwargs)
plt.hist(y, **kwargs)
plt.hist(z, **kwargs)
plt.hist(a, **kwargs)

plt.show()
```

```python
mean = [0,0]
covriance = [[1,1],[1,2]]

x,y = np.random.multivariate_normal(mean, covriance, 100000).T
sample = np.random.multivariate_normal([1,1], covriance, 10000)
```

```python
plt.hist2d(x, y, bins = 30, cmap ='Blues')
plt.colorbar(label ='counts in bins')
plt.title('A 2-d histogram')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.show()
```

```python
plt.hist2d(samples[:,0], samples[:,1], bins = 35, cmap ='Blues')
plt.colorbar(label ='Colorbar')
plt.show()
```
