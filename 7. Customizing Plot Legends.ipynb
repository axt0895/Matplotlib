```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

sns.set_style('ticks')
```

```python
x = np.linspace(0,10,1000)
```

```python
figure, axes = plt.subplots()

axes.plot(x, np.sin(x),'-b', label ='Sine Curve')
axes.plot(x, np.cos(x),'--r', label ='Cosine Curve')

axes.axis('equal')
axes.legend()

plt.show()
```

```python
figure, axes = plt.subplots()

axes.plot(x, np.sin(x),'-b', label ='Sine Curve')
axes.plot(x, np.cos(x),'--r', label ='Cosine Curve')

axes.axis('equal')
axes.legend(loc ='center right', frameon = True, ncol =3)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(x, np.sin(x),'--r', label ='A sine curve')
ax.plot(x, np.cos(x), '-b', label='A cosine curce')
ax.plot(x, np.sin(x+1), ':k', label ='A tane Curve')

ax.legend(loc ='lower center',fancybox = True, shadow = True, borderpad = 1)
plt.show()
```
