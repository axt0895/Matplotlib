```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10, 1000)
```

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x),'-b', label ='Sine Curve')
ax.plot(x, np.cos(x),'--r', label ='Cosine Curve')
ax.axis('equal')
legend = ax.legend(loc ='lower center', ncol = 2)
```
