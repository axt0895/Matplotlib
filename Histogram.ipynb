```python
'''
Histograms are a powerful and commmonly used tools in statistics and
data analysis. They provide a visual representation of the distribution
of the dataset, showing the frequency or count of values within different
intervals or 'bins'. 
'''
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
np.random.RandomState(0)
```

```python
data = np.random.randn(10000)
```

```python
plt.hist(data, bins = 20, color = '#06BFD5', edgecolor ='lightblue', density=True,
        label = 'A histogram plot', orientation = 'vertical')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(' Histogram Example')
plt.legend()
```

```python
#From chatGPT

# Generate some random data for the histogram
data = np.random.randn(1000)

# Plot the histogram
plt.hist(
    data,
    bins=20,            # Number of bins or bin edges
    range=None,         # Range of the data (min and max values)
    density=False,      # Normalize to form a probability density
    cumulative=False,   # Display cumulative distribution
    color='#06BFD5',       # Color of the bars
    edgecolor='lightblue',  # Color of the edges of the bars
    alpha=0.7,          # Transparency of the bars (0.0 to 1.0)
    orientation='vertical',  # Orientation of the bars ('horizontal' or 'vertical')
    log=False,          # Log scale on the y-axis
    label='A histogram Plot'          # Label for the legend
)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
