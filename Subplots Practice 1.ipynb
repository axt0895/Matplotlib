```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(0)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize = (8,4))
fig.set_facecolor('white')

x = np.random.rand(50) 
y = np.random.rand(50)
ax[0,0].scatter(x,y, color ='#2076B3',s =27)
ax[0,0].set_ylim(0,1)
ax[0,0].set_xlim(0,1)
ax[0,0].set_title('Scatter plot')

x1 = np.arange(1,11)
y1 = np.random.randint(0,10,10)
ax[0,1].step(x1, y1)
ax[0,1].set_xlim(0,11)
ax[0,1].set_ylim(0,10)
ax[0,1].set_title('Step plot')

x3 = np.random.randn(1000)
ax[0,2].hist(x3, bins = 15, color = '#2076B3', alpha =0.7)
ax[0,2].set_title('A histogram')


date_rng = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
data = np.random.randn(len(date_rng))
df = pd.DataFrame(data, columns=['Value'], index=date_rng)
ax[1,0].plot(df.index, df['Value'])

x5 = np.random.randn(1000)
ax[1,2].boxplot(x5, vert=False, patch_artist=True)
```
